```python
bowler_df = pd.read_csv(r"C:\Users\rajat\OneDrive\Desktop\IPLData\bowler_stats.csv")
bowler_df.head(10)
```

```python
merged = pd.merge(df, bowler_df, left_on="batsman", right_on="bowler")
merged.head()
```

```python
merged.sort_values(by=["total_runs", "wickets"], ascending=[False, False]).head(10)
```

```python
merged["allrounder_score"] = merged["total_runs"] + (merged["wickets"] * 20)
merged.head()
```

```python
merged.sort_values(by="allrounder_score", ascending=False).head(10)
```

```python
merged["allrounder_score"] = (merged["total_runs"] * 1) + (merged["wickets"] *10)
merged.sort_values(by="allrounder_score", ascending=False).head(10)
    
```

```python
merged.sort_values(by="wickets", ascending=False).head(10)
```

```python
merged["allrounder_score"] = (
    (merged["total_runs"] / merged["total_runs"].max()) * 50 +
    (merged["wickets"] / merged["wickets"].max()) * 50
)

merged.sort_values(by="allrounder_score", ascending=False).head(10)
```

```python
plt.figure()
plt.scatter(merged["total_runs"], merged["wickets"])
plt.xlabel("Total Runs")
plt.ylabel("Wickets")
plt.title("All-rounder Distribtuion")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Features (input)
X = df[["total_runs", "balls_faced", "strike_rate"]]

# Target (output)
y = df["player_type"]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = DecisionTreeClassifier()

# Train
model.fit(X_train, y_train)

# Accuracy
accuracy = model.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
import pandas as pd

new_player = pd.DataFrame([[1000, 800, 150]],
                          columns=["total_runs", "balls_faced", "strike_rate"])

prediction = model.predict(new_player)
print(prediction)
```
